```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/Boston.csv')
```

```python
print('number of rows: ', len(data))
print('columns: ')
data.columns
```

rows: represent different data points.

```python
pd.plotting.scatter_matrix(data)
```

there are a lot of correlations in the data.

```python
col_idx = data.columns.get_loc('crim')
scatter_row = pd.plotting.scatter_matrix(data)[col_idx]
```

```python
# plot all the columns versus crim
_, axes = plt.subplots(nrows=len(data.columns)-1, ncols=1, figsize=(5,5*(len(data.columns)-1)))
ax_num = 0
for i in range(len(data.columns)):
    col = data.columns[i]
    if col == 'crim':
        continue
    axes[ax_num].scatter(data['crim'], data[col])
    axes[ax_num].set(xlabel='crim', ylabel=col)
    ax_num += 1
plt.tight_layout()
plt.show()
```

there is a relationship of crim with other predictors such as medv.
