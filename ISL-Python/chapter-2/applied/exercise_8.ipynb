```python
import pandas as pd
```

```python
college = pd.read_csv("https://www.statlearning.com/s/College.csv")
```

```python
college2 = pd.read_csv("https://www.statlearning.com/s/College.csv", index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
```

```python
college = college3
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
```

```python
college.head()
```

```python
college.boxplot(column="Outstate", by="Private")
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,0.5,1], labels=['No', 'Yes'])
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite')
```

```python
import matplotlib.pyplot as plt
```

```python
college.dtypes
```

```python
plt.subplots(3,2)
plt.subplot(3,2,1)
college['Apps'].plot.hist(bins=10)
plt.xlabel('Apps')
plt.subplot(3,2,2)
college['Accept'].plot.hist(bins=10)
plt.xlabel('Accept')
plt.subplot(3,2,3)
college['Enroll'].plot.hist(bins=10)
plt.xlabel('Enroll')
plt.subplot(3,2,4)
college['Acceptance Rate'] = college['Accept'] / college['Apps']
college['Acceptance Rate'].plot.hist(bins=10)
plt.xlabel('Acceptance Rate')
plt.subplot(3,2,5)
college['Grad.Rate'].plot.hist(bins=10)
plt.xlabel('Grad.Rate')

plt.tight_layout()
plt.show()
```

observation:
most schools have relatively fewer students in applicaitons and accept and enroll.
the acceptance rate of most schools are around 80%.
the graduate rate has a roughly bell shape, and is centered around 70%.
